```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
savings = [0, 100, 150, 200, 300, 500, 300, 350, 500, 600, 750, 1000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', "Oct", 'Nov', 'Dec']



plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Accounts Results')
plt.show
```
